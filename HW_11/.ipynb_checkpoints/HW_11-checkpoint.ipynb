## Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
ratings = pd.read_csv(r'ml-latest-small/ratings.csv')
```

```python
def classification_by_ratings(row):
    if row['rating'] <= 2:
        movie_rating = "низкий рейтинг"
    elif row['rating'] <= 4:
        movie_rating = "средний рейтинг"
    else:
        movie_rating = "высокий рейтинг"
    return movie_rating
```

```python
ratings['class_rating'] = ratings.apply(classification_by_ratings, axis=1)
```

```python
value_counts = ratings['class_rating'].value_counts(dropna=True, sort=True)
value_counts
```

## Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

`geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}`

Результат классификации запишите в отдельный столбец region.

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
keywords = pd.read_csv(r'ml-latest-small/keywords.csv')
```

```python
keywords.head()
```

```python
def geo_classification(row):
    if row['keyword'] in geo_data['Центр']:
        region = "Центр"
    elif row['keyword'] in geo_data['Северо-Запад']:
        region = "Северо-Запад"
    elif row['keyword'] in geo_data['Дальний Восток']:
        region = "Дальний Восток"
    else:
        region = "undefined"
    return region
```

```python
keywords['region'] = keywords.apply(geo_classification, axis=1)
```

```python
value_counts = keywords['region'].value_counts(dropna=True, sort=True)
value_counts
```

## Задание 3 (бонусное)

Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы https://grouplens.org/datasets/movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.


2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

 1. для каждой строки пройдитепо всем годам списка years
 2. если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
 3. если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год


3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’


4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
import re
```

```python
years = list(range(1950, 2011))
```

```python
def production_year(data):
    for year in years:
        if str(year) in data:
            release_date = year
            return release_date
    else:
        release_date = 1900
        return release_date
```

```python
movies = pd.read_csv(r'ml-latest-small/movies.csv')
ratings = pd.read_csv(r'ml-latest-small/ratings.csv').join(movies.set_index('movieId'), on='movieId').drop(['timestamp', 'genres'], axis=1)
ratings['year'] = ratings['title'].apply(production_year)
```

```python
ratings = ratings[ratings.year != 1900]
```

```python
ratings.groupby(['year'])['rating'].mean().sort_values(ascending=False).head()
```
