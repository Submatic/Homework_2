## Домашнее задание к лекции «Случайные события. Случайные величины»

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш (см. справа)

Выгодна ли игра игроку?

Сколько в среднем приобретает или теряет игрок за одну игру?

1. 777: 200 руб.
2. 999: 100 руб.
3. 555: 50 руб.
4. 333: 15 руб.
5. 111: 10 руб.
6. -77: 5 руб.
7. --7: 3 руб.
8. -00: 2 руб.
9. --0: 1 руб.

– любая цифра

```python
import numpy as np
```

```python
trials = 100000    # Количество попыток (потраченных рублей)
gain = 0    # Сумма выйгрыша

for i in range(trials):
    random_score = np.random.randint(1, 1000)    # Случайное число от 1 до 999
    if random_score % 777 == 0:
        gain += 200
    elif random_score  % 999 == 0:
        gain += 100
    elif random_score  % 555 == 0:
        gain += 50
    elif random_score  / 333 == 1:
        gain += 15
    elif random_score  / 111 == 1:
        gain += 10
    elif random_score  % 100 == 77:
        gain += 5
    elif random_score  % 10 == 7:
        gain += 3
    elif random_score  % 100 == 0:
        gain += 2
    elif random_score  % 10 == 0:
        gain += 1
    
        
print(f'Потрачено {trials} рублей, заработано {gain} рублей. \nВ среднем за одну игру игрок теряет {(trials - gain)/trials} рублей. \nЧастота выигрышей по методу Монте-Карло: {round(gain/trials*100)}%.')
```
