### 3. Рассчитайте и визуализировать корреляционную матрицу для количественных переменных.

```python
corrMatrix = df.corr()
corrMatrix
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(corrMatrix, annot=True, linewidths=0.01)
plt.show()
```

```python
# Функция для получения имен колонок с нечисловыми значениями, которые нужно удалить из датасета
def get_redundant_pairs(df):
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

# Функция для получения N первых значений из отсортированного списка значений для коррелирующих пар
def get_top_abs_correlations(df, corr, ascending, n=2):
    labels_to_drop = get_redundant_pairs(df.select_dtypes(include=np.number))
    au_corr = corr.drop(labels=labels_to_drop).sort_values(ascending=ascending)
    return au_corr[0:n]
```

Определите две самые скоррелированные.

```python
au_corr = corrMatrix.abs().unstack()
get_top_abs_correlations(df, au_corr, False)
```

Определите две наименее скоррелированные переменные.

```python
get_top_abs_correlations(df, au_corr, True)
```
